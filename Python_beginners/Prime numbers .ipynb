```python
for i in range(2,101):
    found_prime = True 
    for j in range(2,int(i/2)+1) :
        
        if i%j == 0 :
            print(i,"\t is not prime. ")
            found_prime = False
            break 
        
    if found_prime == True :
        print(i,"\t is prime. ")
        found_prime = True 
        
        
```
